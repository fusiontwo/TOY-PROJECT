```python
from bs4 import BeautifulSoup
# 사이트 정보 가지고 오기
response = driver.page_source
bs = BeautifulSoup(response, 'lxml')
```

```python
bs
```

```python
str(bs.select('title')[0]).strip("<title>").strip("</title>")
```

```python
data = bs.select('div dl dt')
list1 = [str(i).replace("<dt>","") for i in data]
list2 = [str(i).replace("</dt>","") for i in list1]
print(list2)
```

```python
df1 = pd.DataFrame(list2)
df1.columns = ['feature']
df1.index = df1.index + 1
df1
```

```python
data2 = bs.select('div dl dd')
list3 = [str(i).replace("<dd>","") for i in data2]
list4 = [str(i).replace("</dd>", "") for i in list3]
list4
```

```python
# raw data의 패턴이 달라서 추출이 어려운 정보가 7개 존재한다. 바로 다음 코드에서 수정할 부분이다.
df2 = pd.DataFrame(list4)
df2.columns = ['detail']
df2.index = df2.index + 1
df2
```

```python
df2 = pd.DataFrame(list4)
df2.columns = ['detail']
df2.index = df2.index + 1
# 패턴이 달라서 추출하기 어려웠던 데이터 7개는 직접 수정하였다.
df2.loc[1] = ['RefSeq GCF_000003025.6']
df2.loc[2] = ['GenBank GCA_000003025.6']
df2.loc[3] = ['Sus scrofa']
df2.loc[7] = ['AEMK02']
df2.loc[23] = ['SAMN02953785']
df2.loc[29] = ['AEMK']
df2.loc[36] = ['This pig genome sequence (Sscrofa11) has been released by the International Swine Genome Sequencing Consortium under the terms of the Toronto Statement (Nature 2009, 461: 168). The Consortium is coordinating genome-wide analysis, annotation and publication.</p><p class="MuiTypography-root MuiTypography-body1 css-1thjmlm">The sequence data from which this assembly was constructed largely comprise 65x genome coverage in whole genome shotgun (WGS) Pacific Biosciences long reads (Pacific Biosciences RSII, with P6/C4 chemistry). Illumina HiSeq2500 WGS paired-end and mate pair reads were used for final error correction using PILON. Sanger and Oxford Nanopore sequence data from a few CHORI-242 BAC clones were used to fill gaps. All the WGS data were generated from a single Duroc female (TJ Tabasco, also known as Duroc 2-14) which was also the source of DNA for the CHORI-BAC library.</p><p class="MuiTypography-root MuiTypography-body1 css-1thjmlm">Sscrofa11 replaces the previous assembly, Sscrofa10.2, which was largely established from the same Duroc 2-14 DNA source.']
df2
```

```python
final_df = df1["detail"] = df2
final_df
```

```python
final_df = pd.concat([
    pd.DataFrame(df1, columns=['feature']),
    pd.DataFrame(df2, columns=['detail'])
], axis=1)
final_df
```
